# Creating a neural network in pytorch


```python
import torch
import torch.nn as nn ## neural network modules,architecture
import torch.optim as optim ## optimizers 
import torch.nn.functional as F ## activation fns
from torch.utils.data import DataLoader ## to load and use dataset
import torchvision.datasets as datasets
import torchvision.transforms as transforms ## transformations for our datasets

```

```python
### defining the NN class that inherits from nn.Module

class NN(nn.Module):
    def __init__(self,input_size,num_classes):
        super(NN,self).__init__()
        self.fc1 = nn.Linear(input_size,50)
        self.fc2 = nn.Linear(50,num_classes)
    
    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
## create an object
model = NN(784,10)
## trial data
x = torch.randn(64,784)

print(model(x).shape)
```

```python
## set the hyperparameter
input_size = 784
num_classes = 10
learning_rate = 0.001
batch_size = 64
num_epochs = 5
```

```python
### load data
train_dataset = datasets.MNIST(root='dataset/',train=True,transform=transforms.ToTensor(),download=True)
train_loader = DataLoader(dataset=train_dataset,batch_size=batch_size,shuffle=True)

test_dataset = datasets.MNIST(root='dataset/',train=False,transform=transforms.ToTensor(),download=True)
test_loader = DataLoader(dataset=test_dataset,batch_size=batch_size,shuffle=True)
```

```python
## initialize the network

model = NN(input_size=input_size,num_classes=num_classes).to(device)
```

```python
## loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(),lr=learning_rate)
```

```python
## train network
for epoch in range(1):
    for batch_idx,(data,targets) in enumerate(train_loader):
        ## get data to cuda
        data = data.to(device=device)
        targets = targets.to(device=device)
        #print(data)
        print(targets)
        # get to correct shape

        data = data.reshape(data.shape[0],-1)


        # forward
        scores = model(data)
        loss = criterion(scores,targets)
        
        ## backward
        ## set grad zero for each batch
        optimizer.zero_grad()
        loss.backward()
        
        ## gradient descent
        optimizer.step()
        
        
def check_accuracy(loader,model):
    num_correct = 0
    num_samples = 0
    model.eval()
    
    ## no grad computation for this calculation
    with torch.no_grad():
        for x,y in loader:
            x=x.to(device=device)
            y=y.to(device=device)
            x = x.reshape(x.shape[0],-1)
            
            scores = model(x)
            _,predictions = scores.max(1)
            
            num_correct+= (predictions==y).sum()
            num_samples+=predictions.size(0)
            
        print(num_correct/num_samples)
    model.train()


check_accuracy(train_loader,model)
check_accuracy(test_loader,model)
```

### Loading and getting the data
* Dataload class (to load the data into {data,label}) dictionary
* One can use dataloader object as an iterator

```python
x=np.arange(100)
y=[[j for i in range(10)] for j in range(10)]
y=np.array(y).flatten()
data = pd.DataFrame({'x':x,'y':y})
print(f'{data.head()}')
print(f'{data.tail()}')
```

```python
### class to load the data set 
### the output data can be 
class Dataload(Dataset):
    def __init__(self, data,device=device):
        self.x_data = torch.tensor(data.iloc[:,0].values,dtype=torch.float32).to(device=device)
        self.y_data = torch.tensor(data.iloc[:, 1].values, dtype=torch.float32).to(device=device)

    def __len__(self):
        return len(self.y_data)

    def __getitem__(self,idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()
        
        image = self.x_data[idx]
        label = self.y_data[idx]
        return {'data':image,'lab':label}

### creating a dataloader (iterator) and data set 

data_ld = Dataload(data)
tr_dloader = DataLoader(data_ld,batch_size=10,shuffle=True)
for epoch in range(1):
    for (batch_idx,batch) in enumerate(tr_dloader):
        print(batch['data'],batch['lab'])
```
